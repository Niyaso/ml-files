### Split data into X and y

```python
X = data[['TV','radio','newspaper']]

X.head()
```

```python
#Split the data into X and y
X = data.drop("sales",axis=1)

y = data['sales']
```

### Split data for training and testing

```python
#split the data for trai4ning and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25) #0.75 data will be for training.

```

```python
X_train.shape
```

```python
X_train.head()
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Load and create Linear Regression object for training

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression() #creating object of LinearRegression

reg.fit(X_train,y_train) #training and fitting LR object using training data

#fit command is used for training a model using train data.
```

### Evaluation of Coefficients  and Y-intercepts (C)

```python
#coeeficients of liner regression - m1,m2,m3
reg.coef_
```

```python
#intercept of liner regression - c
reg.intercept_
```

### The Linear Equation will be sales = 0.04543596*TV + 0.19147751*radio + 0.00254094*newspaper + 2.7783929910850524

### Predict value for X-test

```python
ypred=reg.predict(X_test)  #predicting the sales for testing data

ypred
```

### Calculating MSE

```python
from sklearn.metrics import mean_squared_error #calculating MSE

MSE=mean_squared_error(ypred,y_test)
print("MSE:",MSE)
```

```python
sum((y_test-ypred)**2)/len(ypred) #this is how we can calculate MSE.
```

```python
Results = X_test.copy()
Results['Actual']=y_test
Results['Predicted']=ypred
Results.head()
```

```python
sns.lineplot(x='TV',y='Actual',data=Results)
sns.lineplot(x='TV',y='Predicted',data=Results)
plt.plot()
plt.title('actual vs predicted')
plt.xlabel('TV')
plt.ylabel('actual vs predicted')
plt.show()
```

```python
sns.lineplot(x='radio',y='Actual',data=Results)
sns.lineplot(x='radio',y='Predicted',data=Results)
plt.plot()
plt.title('actual vs predicted')
plt.xlabel('Radio')
plt.ylabel('actual vs predicted')
plt.show()
```

```python
sns.lineplot(x='newspaper',y='Actual',data=Results)
sns.lineplot(x='newspaper',y='Predicted',data=Results)
plt.plot()
plt.title('actual vs predicted')
plt.xlabel('Newspaper')
plt.ylabel('actual vs predicted')
plt.show()
```
