```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv('/Users/niyasop/Desktop/ml-packagess/carprices.csv')
```

```python
df
```

```python
df.shape
```

```python
plt.scatter(df['Car Model'],df['Sell Price($)'],color='red',alpha=0.6)
plt.title('car model vs sell price',color='green')
plt.xlabel('car model')
plt.ylabel('sell price')
plt.show()
```

```python
df
```

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final=merged.drop(['Car Model','Audi A5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop(['Sell Price($)'],axis='columns')
```

```python
x
```

```python
y=(final['Sell Price($)'])
y
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
